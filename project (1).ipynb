# Task 3: What are the Most Common Job Titles?

```python
most_common_jobs = df["job_title"].value_counts().head(10)
most_common_jobs
```

```python
top_jobs = df["job_title"].value_counts().head(15)

plt.figure(figsize=(10,6))
top_jobs.sort_values().plot(kind="barh")

plt.title("Top 10 Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.savefig("top_job_titles.png")
plt.show()
```

### Analysis
Data Scientist appears as the most common role in the dataset.

# Task 4: How Does Experience Affect Salary?

```python
experience_salary = df.groupby("experience_level")["salary_in_usd"].mean()
experience_salary
```

```python
experience_salary.plot(kind="bar")

plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.savefig("salary_by_experience.png")
plt.show()
```

### Analysis
Entry-level professionals earn the lowest salaries, while Executive-level professionals earn the highest compensation packages.

# Task 5: What is the Salary Distribution?

```python
plt.hist(df["salary_in_usd"], bins=10)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.savefig("salary_distribution.png")
plt.show()
```

# Task 6: Which Employment Type is Most Common?

```python
emp_type = df["employment_type"].value_counts()
emp_type
```

```python
emp_type.plot(kind="bar")

plt.title("Employment Type Distribution")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.savefig("employment_type_distribution.png")
plt.show()
```

### Analysis
Full-Time employment dominates the Data Science industry.

# Task 7: How Popular is Remote Work?

```python
remote = df["remote_ratio"].value_counts()
remote
```

```python
plt.pie(remote,labels=remote.index,autopct="%1.1f%%")
plt.title("Remote Work Distribution")
plt.savefig("remote_work_distribution.png")
plt.show()
```

### Analysis
Data Science is one of the industries most adaptable to remote work environments.

# Task 8: Does Remote Work Influence Salary?

```python
remote_salary = df.groupby("remote_ratio")["salary_in_usd"].mean()
remote_salary
```

```python
remote_salary.plot(kind="bar")

plt.title("Salary by Remote Ratio")
plt.xlabel("Remote Ratio")
plt.ylabel("Average Salary")
plt.savefig("salary_by_remote_ratio.png")
plt.show()
```

### Analysis
Employees working fully remotely tend to earn higher average salaries.

# Task 9: What is the Distribution of Company Sizes?

```python
company_size = df["company_size"].value_counts()
company_size
```

```python
plt.pie(company_size,labels=company_size.index,autopct="%1.1f%%")

plt.title("Company Size Distribution")
plt.savefig("company_size_distribution.png")
plt.show()
```

### Analysis
Medium-sized companies represent the largest portion of employers.

# Task 10: Does Company Size Affect Salary?
This analysis evaluates the relationship between company size and average employee salary.

```python
size_salary = df.groupby("company_size")["salary_in_usd"].mean()
size_salary
```

```python
size_salary.plot(kind="bar")

plt.title("Salary by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Average Salary")
plt.savefig("salary_by_company_size.png")
plt.show()
```

### Analysis
Large companies generally offer higher salaries than medium and small companies.

# Task 11: Top Employee Residence Countries

```python
countries = df["employee_residence"].value_counts().head(10)
countries
```

```python
countries.plot(kind="barh")

plt.title("Top Employee Countries")
plt.xlabel("Count")
plt.ylabel("Country")
plt.savefig("top_employee_countries.png")
plt.show()
```

### Analysis
The United States contributes the highest number of Data Science professionals.

# Task 12: Salary Trend by Year

```python
year_salary = df.groupby("work_year")["salary_in_usd"].mean()
year_salary
```

```python
avg_salary_year = df.groupby("work_year")["salary_in_usd"].mean()
avg_salary_year.plot(kind="bar")

plt.title("Average Salary by Year")
plt.xlabel("Year")
plt.ylabel("Average Salary")

plt.xticks(rotation=0)
plt.savefig("salary_trend_by_year.png")
plt.show()
```

### Analysis
Average salaries increased consistently over the years, reflecting the growing demand and value of Data Science skills worldwide.
